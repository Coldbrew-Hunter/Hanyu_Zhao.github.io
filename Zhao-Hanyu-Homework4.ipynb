```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import ipywidgets
```

```python
ufo = pd.read_csv("https://github.com/UIUC-iSchool-DataViz/is445_AOUAOG_fall2021/raw/master/week05/data/ufo-scrubbed-geocoded-time-standardized-00.csv",
                  names = ["date", "city", "state", "country",
                          "shape", "duration_seconds", "duration",
                          "comment", "report_date", "latitude", "longitude"],
                  parse_dates = ["date", "report_date"])
```

```python
axis_names_subset = ['date', 'state', 'country', 'shape',
                     'duration_seconds', 'report_date', 'latitude', 'longitude']
```

```python
%matplotlib inline
@ipywidgets.interact(xname = axis_names_subset, yname = axis_names_subset,cmap = plt.colormaps())
def ufo_scatter(xname,yname,cmap):
        fig = plt.figure(figsize=(10,5))
        x = ufo[xname]
        y = ufo[yname]
        plt.scatter(x,y,c=np.log10(ufo['duration_seconds']),cmap = cmap)

        plt.show()
```

```python
categorical_columns = ['state', 'country', 'shape', 'duration_seconds_log', 'latitude']
ufo['duration_seconds_log'] = np.log10(ufo['duration_seconds'])
```

```python
@ipywidgets.interact(bin_name = categorical_columns, nbins = np.arange(1,10))
def make_scatter(bin_name, nbins):
    fig, ax = plt.subplots(figsize=(20, 15))
    myseries = ufo.groupby(bin_name)[bin_name].count()
    if type(ufo[bin_name]) != float and type(ufo[bin_name]) != int:
            ax.bar(myseries.index, myseries.values)
    else: 
        ufo.hist(column=bin_name, ax=ax, bins=nbins)
    fig.autofmt_xdate(rotation=90)
    plt.show()
```
